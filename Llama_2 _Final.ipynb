```python
# 13 billion model
model_name_or_path = "TheBloke/Llama-2-13B-chat-GGML"
model_basename = "llama-2-13b-chat.ggmlv3.q5_1.bin"
```

```python
from llama_cpp import Llama
from huggingface_hub import hf_hub_download
```

```python
model_path = hf_hub_download(repo_id=model_name_or_path, filename=model_basename)
# GPU
lcpp_llm = None
lcpp_llm = Llama(
    model_path=model_path,
    n_threads=2, # CPU cores
    n_batch=8192, # Should be between 1 and n_ctx, consider the amount of VRAM in your GPU.
    n_gpu_layers=32,
    n_ctx=1100
    #n_gqa=8# Change this value based on your model and your GPU VRAM pool.
    )
```

```python
prompt = """
Reliance's jamnagar refinery with a capacity of 2.4 million barrels per day, due to fire in a cracking unit the output is reduce to 30%.
 this happened on september 20, 2023 and it expected to take a week to repair the damage.

"""
prompt_template=f'''SYSTEM:As an energy analyst, I want model's help in storing the value in

     JSON data format or nested JSON data format accordingly of all the important information

      such as the refinery name, affected unit, affected capacity, shutdown date and restart date etc.

    from the given.

USER: {prompt}

ASSISTANT:
'''

response=lcpp_llm(prompt=prompt_template, max_tokens=-1, temperature=0.0, top_p=0.1,
                  echo=False)
#print(response)
print(response["choices"][0]["text"])
```
